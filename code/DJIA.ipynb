```python
get_ipython().magic('matplotlib inline')

import logging
logger = logging.getLogger()
logger.setLevel(logging.DEBUG)
logging.debug("test")

from pylab import *
from pandas.io.data import *
from pandas import *
from pandas.tools.plotting import *

class DJIASmallModel:
    symbols = ['IBM', 'MMM', 'CVX', 'UTX', 'XOM'] # List of all stock symbols to download #'JNJ', 'MCD'

    def __init__(self):
        self.__download_data()
        self.__extract_data()

    def __download_data(self):
        logging.debug('Reading stock data from Yahoo Finance')
        # extra half a year to warm EWMA up for historical vol calc
        self.stock_data = DataReader(self.symbols, 'yahoo', start='07/01/2009',end='06/30/2012') # Download data from YAHOO as a pandas Panel object
    
    def __extract_data(self):
        logging.debug('Extracting and plotting Adjusted Closing Prices')
        self.adj_close  = self.stock_data['Adj Close']                  # Pull out adjusted closing prices as pandas DataFrane object
        self.adj_close = concat([self.adj_close,DataFrame(sum(self.adj_close, axis=1))], axis=1)
        #self.adj_close.plot(title='Daily Adjusted Closing Prices for 401K Funds')
        #show()
        self.returns = self.adj_close.pct_change()
        #self.returns.plot(title='Arithmetric Daily Returns')
        #show()
        #plot(rolling_std(returns.IBM, window=25))
        #show()
        #plot(ewma(returns**2, com=0.94/(1-0.94)))
        self.ewma_cor = ewma(self.returns**2, com=0.94/(1-0.94))

        # realign the data properly to start from 2010/01/01
        self.ewma_cor = self.ewma_cor["2010/01/01":]
        self.returns = self.returns["2010/01/01":]

        logging.debug(self.ewma_cor.shape)
        logging.debug(self.ewma_cor[1:5])
        logging.debug(self.returns.shape)
        logging.debug(self.returns[1:5])
        #ewmcorr(arg1=returns, arg2=returns, com=0.94/(1-0.94), pairwise = True)
        #plot(-sqrt(ewma(returns.IBM**2, com=0.94/(1-0.94))))
        
    def __plot_extra():
        """
        Create a scatter plot of mean return vs. standard deviation (a measure of volatility)
        for each stock. We would expect that the mean return would increase with volatility,
        but this is not always the case.
        """
        mean_ret   = returns.mean() # Returns a pandas DataFrame with mean values indexed by stock symbol
        std_ret    = returns.std()  # Returns a pandas DataFrame with standard deviations (volatility) indexed by stock symbol
        plot(std_ret,mean_ret,'ro')
        for ii in range(len(symbols)):
            annotate(mean_ret.index[ii],(std_ret[ii],mean_ret[ii])) # Annotate the markers with the stock symbol
        title('Mean Daily Return vs. Standard Deviation for 401K Funds')
        xlabel('Standard Deviation of Daily Returns')
        ylabel('Mean of Daily Returns')
        show()
        """
        Now create a boxplot as another means to visualize volatility in a stock. Note that the 
        boxplot gets pretty busy for more than a handful of stocks
        """

        returns.boxplot()
        title('Selected 401K Funds Daily Returns')
        show()

        """
        Finally calculate the correlation matrix between the various stocks. This returns a DataFrame
        object with rows and columns labeled with the stock symbols.  It's actually fairly clever
        because it automatically handles the fact that not all the symbols have data available on the
        same day.
        The values method returns just the array of data
        """

        sel_corr=returns.corr()     # Calculate correlation
        sel_corr_values=sel_corr.values # Return just the array of correlation data (no labels)
        print(sel_corr_values)
    
        """
        Now create a scatter plot matrix. This shows a probability density on each diagonal and
        a scatter plot indicating correlation on the off-diagonals
        """

        scatter_matrix(returns, alpha=0.2, figsize=(6,6), diagonal='kde')
        show()

djia = DJIASmallModel()
```

```python
# model setup
u_dim = len(djia.symbols)

def A(u):
    return np.eye(u_dim)

def C(u):
    return np.eye(u_dim)

def B(u):
    return np.eye(u_dim)

def D(u):
    return np.eye(u_dim)

def F(u): # need to return a vector
    return u

def G(u): # need to return a vector
    return np.zeros(u_dim)

R = eye(n_u)
Q = 0.0001*eye(n_u)
L = 0.0001*eye(n_u)
    
def gamma_10(t):
    return 10

def gamma_100(t):
    return 100

def gamma_t(t):
    return t

def gamma_increasing(t):
    return min(100,10*t)

# proposal distribution (bootstrapping proposal)
def sample_proposal_u(u_old, gamma=1):
    """ U_t | U_t-1 """
    return stats.multivariate_normal(zeros(u_dim), linalg.inv(L*gamma)).rvs()

def density_proposal_u(u, gamma=1):
    return exp(-0.5*u.dot(L*gamma).dot(u.T))

def likelihood(y, mu, var, gamma=1):
    y1 = y-IM
    return exp(-0.5*gamma*y1.dot(linalg.inv(var)).dot(y1.T))

# resample from data array
def resample(data, indices, fromIndex, toIndexExclusive):
    data[fromIndex:toIndexExclusive,:] = data[fromIndex:toIndexExclusive,indices]

def kf_step(X, P, A, Q, F, Y, C, R, G):
    X = np.dot(A, X) + F
    P = np.dot(A, np.dot(P, A.T)) + Q # Q=BB'
    IM = np.dot(C, X) + G # it is empty
    IS = np.dot(C, np.dot(P, C.T)) + R # R=DD'
    K = np.dot(P, np.dot(C.T, linalg.inv(IS)))
    X = X + np.dot(K, (Y-IM))
    P = P - np.dot(K, np.dot(IS, K.T))
    return (X,P,K,IM,IS)

def cost(IS, IM, y, u, t):
    a = likelihood(y, IM, IS, t)
    b = density_proposal_u(u, t)
    return(a,b,a)
```

```python
mcmc_setting=[0]
TT = [30]
NN = [1000]
reps = 3

res = dict()

for setting in itertools.product(mcmc_setting, TT, NN):
    print("Setting: ", setting)   
    (mcmc_on, T, N) = setting

    ########################### Ref model setup
    y_ref = u_dim*djia.returns.values[:,-1] # make it the sum instead of average
    T = len(y_ref)

    x_sigma = djia.ewma_cor.values[:,:-1] # ewma covariance valuet
    x_Sigma = zeros(shape=(T, u_dim, u_dim)) # covariance matrix [time, dim x, dim x]
    for i in range(len(x_sigma)):
        x_Sigma[i] = diag(x_sigma[i])
    # truncate everything down to T step only
    y = y_ref[0:T].reshape(T,1)
    x_Sigma = x_Sigma[:T]

    ############################ Repeat reps time   
    for rep in range(reps):
        seed = rep*7 # prime number multiple
        np.random.seed(seed)        
        print("Random seed: ", seed)
        
        ################### Initialisation storage
        u = zeros(shape=(T, N, n_u))
        
        cost1 = zeros(shape=(T,N))
        cost2 = zeros(shape=(T,N))
        
        mu = zeros(shape=(T,N, n_u))
        Sigma = zeros(shape=(T,N, n_u, n_u))
        weight = zeros(shape=N)
        
        ################### Main loop
        # timing
        start_time = timeit.default_timer()

        for t in range(T): 
            if t == 0:
                for i in range(N):
                    u[t,i] = 0
                    (mu[t,i],Sigma[t,i],K,IM,IS) = (0, 0, A(0), Q, F(0), y[0], C)
                    (cost1[t,i], cost2[t,i], weight[i]) = cost(IS,IM,y[t,:],u[t,i], t+1) 
            else:
                for i in range(N):
                    u[t,i] = sample_proposal_u(u[t-1,i], t+1)
                    # assume mu[-1] and sigma[-1] are zero (i.e., the last ind)
                    (mu[t,i],Sigma[t,i],K,IM,IS) = kf_step(mu[t-1,i], Sigma[t-1,i], A(u[t,i]), Q, F(u[t,i]), y[t], C(u[t,i]), R)
                    (cost1[t,i], cost2[t,i], weight[i]) = cost(IS,IM,y[t,:],u[t,i], t+1)  
                
            # resampling
            weight = weight/sum(weight)         
            indices = np.random.choice(range(N),size=N, replace=True, p=weight)
            for data in (u, cost1, cost2, mu, Sigma):
                resample(data, indices, 0, t+1)

        elapse_time = timeit.default_timer() - start_time
        
        ##################### store result
        # mode
        i = np.nonzero(np.prod(cost1*cost2, 0) == max(np.prod(cost1*cost2, 0 )))
        uhat.mode = u[:,i[0][0], :]
        uhat.mean = np.mean(u, axis=1)
```
