```python
get_ipython().magic('matplotlib inline')
# import logging

import logging
logger = logging.getLogger()
logger.setLevel(logging.DEBUG)
logging.debug("test")

# import libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats, linalg
from numpy import zeros, ones, eye, identity, cos, sqrt, exp, sum, prod, arange, pi, diag
from numpy.linalg import inv
import itertools
import timeit
import pickle

number_sources = 4
T = 50
x0 = zeros((number_sources, 1))

gamma=3*arange(1,T+1)

# matrices at cost
L2=.1*diag(arange(1,number_sources+1))               # switch on and off action penalty
L1=.1*diag(arange(1,number_sources+1)[::-1])  # continuous action penalty
constraints=2*arange(1,number_sources+1).reshape(1,number_sources)
Ksi=1

R = eye(1)
Q = eye(number_sources)
a = eye(number_sources)
F = eye(number_sources)

n_x = n_u = number_sources
n_y = 1
y = zeros((T,1, n_y))
stepsize=5
delta=4
for t in range(0,T+1,stepsize):
    if t >= 25:
        delta = delta-2
    y[t:,] = y[t:,:] + delta

# resample helper function given the index
def resample(data, indices, fromIndex, toIndexExclusive):
    data[fromIndex:toIndexExclusive,:] = data[fromIndex:toIndexExclusive,indices]
```

```python
reps = 3
N = 100


results = dict()   
for rep in range(reps):
    I = zeros(shape=(T, N))
    
    u1 = ones(shape=(T, N, 1, number_sources)) # make sure they are matrix after all
    u2 = ones(shape=(T, N, 1, number_sources))
     
    rec_true = zeros(shape=(T,N)) # holding the cost
    rec2 = zeros(shape=(T,N))
    

    RecLikeMean=zeros(shape=(T, N, 1, n_y))
    RecLikeVar=zeros(shape=(T,N, 1, n_y))

    # KF
    mu=zeros(shape=(T,N, 1, n_x))
    Sigma=zeros(shape=(T,N, 1, n_x, n_x))

    SigmaInit=zeros(shape=(n_x, n_x))

    weight=zeros(shape=N)
        
    for t in range(T):
        epsilon=gamma[t]
    
        if t==0:
            for i in range(N):
                u2[t, i] = stats.bernoulli.rvs(p=0.5, size=[1, n_u])
                u1[t, i] = u2[t, i]*stats.norm.rvs(size=[1, n_u])*constraints
                
                weight[i] = 1
                weight[i] = exp(-0.5*epsilon*(u1[t,i].dot(L1).dot(u1[t,i].T))).item()
                weight[i] = weight[i] * exp(-0.5*epsilon*(u2[t,i].dot(L2).dot(u2[t,i].T))).item()
                
                rec2[t,i] = weight[i]
        else:
            for i in range(N):
                u2[t, i] = stats.bernoulli.rvs(p=0.5, size=[1, n_u])
                u1[t, i] = u2[t, i]*stats.norm.rvs(size=[1, n_u])*constraints
                
                weight[i] = 1
                weight[i] = exp(-0.5*epsilon*(u1[t,i].dot(L1).dot(u1[t,i].T))).item()
                weight[i] = weight[i] * exp(-0.5*epsilon*((u2[t,i]-u2[t-1,i]).dot(L2).dot((u2[t,i]-u2[t-1,i]).T))).item()
                
                rec2[t,i] = weight[i]
                
        for i in range(N):
            c=u2[t,i]
            a=diag(c[0])
            F=a
            b=a
            
            if t==0:
                mu_pred = a.dot(x0) + u1[t,i].T
                SigmaPred=a.dot(SigmaInit).dot(a)+b.dot(b.T)
            else:
                mu_pred = a.dot(mu[t-1,i].T) + F.dot(u1[t,i].T)
                SigmaPred=a.dot(SigmaInit).dot(a.T)+b.dot(b.T)
            
            z = y[t].T - c.dot(mu_pred)
            SS = c.dot(SigmaPred).dot(c.T) + R
            
            RecLikeMean[t,i] = c.dot(mu_pred)
            RecLikeVar[t,i] = SS
            
            rec_true[t,i] = exp(-0.5*epsilon*(y[t] - RecLikeMean[t,i]).dot(linalg.inv(SS)).dot((y[t] -RecLikeMean[t,i]).T))

            weight[i] = weight[i] * rec_true[t,i]
            
            K=SigmaPred.dot(c.T).dot(linalg.inv(SS))
            Sigma[t,i] = (eye(n_u)-K.dot(c)).dot(SigmaPred)
            mu[t,i] = (mu_pred + K.dot(z)).T

        weight=weight/sum(weight)

        indices = np.random.choice(range(N),size=N, replace=True, p=weight)
        for data in (u1, u2, RecLikeMean, RecLikeVar, mu, Sigma, rec_true, rec2):
            resample(data, indices, 0, t+1)
            
    # alternative calculation? normalised by gamma properly
    gamma_m = zeros(shape=(T,N))
    for t in range(T):
        gamma_m[t,:] = gamma[t]

    j_opt = np.sum(np.log(rec_true*rec2)/gamma_m, 0)
    j = np.max(j_opt)
    j_int = np.nonzero(j_opt == j)[0][0]

    ymean = RecLikeMean[:,j_int]
    yvar = RecLikeVar[:,j_int]
    results[rep] = [rec_true, rec2, mu, Sigma, j_opt, j, j_int, ymean, yvar]
```
